```python
# Numeric Types
# Integer: 
print("Integer:")
print(3)
print(30)
print(402340)

print("\nFloat: ")
# Floats:
print(3.3)
print(2.545)
print(4.5)
# ...

# It does not really matter for python - it will find the proper format for numerical types in 99% of the cases
```

```python
# Waht can we do with numerical types?
# Basic Calculation:

print(3 + 3)      # Addition
print(3 - 5)      # Subtraction
print(3 * 10)     # Multiplication
print(50 / 20)    # Division
print(2 ** 3)     # Exponentiation
print(4 % 2)      # Modulus
print(5 % 2)
```

```python
# Strings - Text Type

'Your Username is: ...' 
print('Please enter a correct password!')
print("Thank you for participating!")
```

```python
# Boolean Comparisons: True or False
print(True)
print(False)
```

```python
# Output is True, IF something is logically correct:
print(3 == 3)
print( 3*3 == 9)
print("password" == "password")

print(2 > 1)
print(2 < 3)
```

```python

# Output is False, IF something is logically incorrect:
print(3 == 4)
print( 3*3 == 10)
print("password" == "Password")

print(2 < 1)
print(2 > 3)
```

```python
# List
print(["Markus", "Student", "markus@univie.ac.at"])
print(["Markus", 23, 180.3, True])
```
