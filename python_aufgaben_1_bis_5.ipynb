3. das Produkt berechnen

## Aufgabe 3: Strings und bool

Schreibe ein Programm, das:

1. deinen Vornamen in einer Variable speichert
2. deinen Nachnamen in einer zweiten Variable speichert
3. beide zu einem vollständigen Namen zusammenfügt
4. eine `bool`-Variable anlegt, die speichert, ob dein Vorname länger als 5 Zeichen ist
5. zusätzlich dein Alter speichert und prüft, ob du volljährig bist

Gib den vollständigen Namen und die booleschen Ergebnisse aus.


```python
# Lösung Aufgabe 3
first_name = "Marcel"
last_name = "Muster"
full_name = first_name + " " + last_name
first_name_longer_than_five = len(first_name) > 5
age_for_check = 28
is_adult = age_for_check >= 18

print(full_name)
print(first_name_longer_than_five)
print(is_adult)
```

## Aufgabe 4: Listen mit Zahlen auswerten

Schreibe ein Programm mit einer `list` von mindestens 5 Zahlen.

Das Programm soll:

1. die Liste ausgeben
2. die erste Zahl ausgeben
3. die letzte Zahl ausgeben
4. die Summe von mindestens 3 Elementen aus der Liste berechnen
5. den Durchschnitt von allen Zahlen berechnen
6. eine neue Zahl zur Liste hinzufügen
7. mit einer `bool`-Variable prüfen, ob mehr als 5 Elemente in der Liste sind


```python
# Lösung Aufgabe 4
numbers = [4, 8, 15, 16, 23]

print(numbers)
print(numbers[0])
print(numbers[-1])

sum_three = numbers[0] + numbers[1] + numbers[2]
print(sum_three)

average = sum(numbers) / len(numbers)
print(average)

numbers.append(42)
print(numbers)

more_than_five = len(numbers) > 5
print(more_than_five)
```

## Aufgabe 5: Dictionary mit gemischten Datentypen

Schreibe ein Programm, das ein `dict` für eine Person erstellt.

Das Wörterbuch soll mindestens diese Informationen enthalten:

- Name (`string`)
- Alter (`int`)
- Größe (`float`)
- mag_python (`bool`)
- Lieblingszahlen (`list` mit mindestens 3 Zahlen)

Danach soll dein Programm:

1. das gesamte Wörterbuch ausgeben
2. nur den Namen ausgeben
3. nur die Lieblingszahlen ausgeben
4. den Durchschnitt der Lieblingszahlen berechnen
5. eine neue Information zum Wörterbuch hinzufügen, zum Beispiel `stadt`
6. optional ein zweites Wörterbuch erstellen und das Alter vergleichen


```python
# Lösung Aufgabe 5
person = {
    "name": "Marcel",
    "age": 28,
    "height": 1.82,
    "mag_python": True,
    "lieblingszahlen": [3, 7, 21]
}

print(person)
print(person["name"])
print(person["lieblingszahlen"])

average_favorite_numbers = sum(person["lieblingszahlen"]) / len(person["lieblingszahlen"])
print(average_favorite_numbers)

person["stadt"] = "Wien"
print(person)

person_2 = {
    "name": "Alex",
    "age": 31,
    "height": 1.75,
    "mag_python": False,
    "lieblingszahlen": [5, 10, 15]
}

print(person["age"] > person_2["age"])
```
