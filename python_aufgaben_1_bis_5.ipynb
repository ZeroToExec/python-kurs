```python
# Lösung Aufgabe 2
training_day_1 = 30
training_day_2 = 45

summe = training_day_1 + training_day_2
differenz = training_day_2 - training_day_1
produkt = training_day_1 * training_day_2
division = training_day_2 / training_day_1

print(summe)
print(differenz)
print(produkt)
print(division)

hours_1 = 1.5
hours_2 = 2.0

print(hours_1 + hours_2)
print(hours_2 - hours_1)
print(hours_1 * hours_2)
print(hours_2 / hours_1)
```

## Aufgabe 4: Herzfrequenzen in einer Liste

Erstelle eine `list` mit mindestens 5 Herzfrequenz-Werten aus einem Training.

Das Programm soll:

1. die ganze Liste ausgeben
2. den ersten Wert ausgeben
3. den letzten Wert ausgeben
4. die Summe von 3 Werten berechnen
5. den Durchschnitt aller Werte berechnen
6. einen neuen Wert zur Liste hinzufügen
7. mit `bool` prüfen, ob die Liste mehr als 5 Werte enthält


```python
# Lösung Aufgabe 4
heart_rates = [92, 98, 105, 110, 102]

print(heart_rates)
print(heart_rates[0])
print(heart_rates[-1])

sum_three = heart_rates[0] + heart_rates[1] + heart_rates[2]
print(sum_three)

average = sum(heart_rates) / len(heart_rates)
print(average)

heart_rates.append(108)
print(heart_rates)

more_than_five = len(heart_rates) > 5
print(more_than_five)
```

## Aufgabe 5: Sportprofil mit Dictionary

Erstelle ein `dict` für eine Person aus dem Sportbereich.

Das Dictionary soll mindestens enthalten:

- Name (`string`)
- Alter (`int`)
- Größe (`float`)
- macht_pilates (`bool`)
- Lieblingsübungen (`list`)

Danach soll dein Programm:

1. das ganze Dictionary ausgeben
2. nur den Namen ausgeben
3. nur die Lieblingsübungen ausgeben
4. zählen, wie viele Lieblingsübungen gespeichert sind
5. eine neue Information hinzufügen, zum Beispiel `sportart`
6. optional ein zweites Dictionary erstellen und das Alter vergleichen


```python
# Lösung Aufgabe 5
person = {
    "name": "Lena",
    "age": 24,
    "height": 1.70,
    "macht_pilates": True,
    "lieblingsuebungen": ["Roll Up", "Bridge", "Plank"]
}

print(person)
print(person["name"])
print(person["lieblingsuebungen"])
print(len(person["lieblingsuebungen"]))

person["sportart"] = "Pilates"
print(person)

person_2 = {
    "name": "Mia",
    "age": 27,
    "height": 1.68,
    "macht_pilates": False,
    "lieblingsuebungen": ["Squat", "Lunge", "Side Plank"]
}

print(person["age"] < person_2["age"])
```
